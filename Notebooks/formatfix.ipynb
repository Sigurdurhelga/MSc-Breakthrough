I'm splitting this all up by | and spaces to get the values



```python
df = pd.read_csv("trimmed2.csv",index_col='fen')
```

```python
col_splits = {}
for col in df.columns:
    print(col)
    col_splits[col] = [list(filter(lambda y: y != '' , x)) for x in df[col].str.split('\|| ').tolist()]
```

```python
all_rows = []
print(len(df))
for i in range(len(df)):
    if i % 1000 == 0:
        print(i)
    row = [df.iloc[i].name]
    for col in df.columns:
        row.extend(col_splits[col][i])
    all_rows.append(row)
```

```python
newcolswhitemg = [col+"WhiteMG" for col in df.columns]
newcolswhiteeg = [col+"WhiteEG" for col in df.columns]
newcolsblackmg = [col+"BlackMG" for col in df.columns]
newcolsblackeg = [col+"BlackEG" for col in df.columns]
newcolstotalmg = [col+"TotalMG" for col in df.columns]
newcolstotaleg = [col+"TotalEG" for col in df.columns]
cols = ["fen"]
for i in range(len(df.columns)):
    cols.append(newcolswhitemg[i])
    cols.append(newcolswhiteeg[i])
    cols.append(newcolsblackmg[i])
    cols.append(newcolsblackeg[i])
    cols.append(newcolstotalmg[i])
    cols.append(newcolstotaleg[i])
ndf = pd.DataFrame(all_rows, columns=cols)
```

```python
ndf.to_csv("data_cs.csv",columns=cols,index=True)
```
